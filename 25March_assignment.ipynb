Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may
impact your analysis.

Ans: Rating column contains 1474  missing value.

Impact of missing datasets:

1) Bias in Descriptive Statistics: Missing values can lead to biased descriptive statistics (mean, median, standard deviation, etc.). If you simply ignore missing values when calculating these statistics, you may get an inaccurate representation of the data.

2) Loss of Information: When missing values are removed or ignored, you lose potentially valuable information, which can affect the quality of your analysis. Removing entire rows with missing values can significantly reduce the size of your dataset.

3) Reduced Statistical Power: Missing values can reduce the statistical power of your analysis. This means that you may be less likely to detect significant relationships or differences in your data because you have less information to work with.

4) Modeling Challenges: Many machine learning algorithms cannot handle missing values directly. You may need to impute missing values or use specialized techniques to work with incomplete data, which can complicate the modeling process.

5) Biased Inferences: If the missing data is not completely random (i.e., it has a pattern or is related to other variables), it can introduce bias into your analysis. For example, if missing values are more likely to occur for certain demographic groups, your analysis may not accurately represent the entire population.

6) False Assumptions: Missing values can violate the assumptions of some statistical tests and models. For example, linear regression assumes that there are no missing values and that the relationship between variables is linear. Violating these assumptions can lead to incorrect conclusions.

7) Imputation Uncertainty: When you impute missing values, you introduce uncertainty into your analysis. Different imputation methods can yield different results, and the choice of imputation technique should be carefully considered.

8) Complexity: Handling missing values often adds complexity to the data preprocessing phase of your analysis. You need to decide on an appropriate imputation method or strategy, which may require domain knowledge or experimentation.

9) Increased Risk of Overfitting: Imputing missing values can lead to overfitting if not done carefully. Imputed values may introduce noise into the data, which can lead to models that fit the noise rather than the underlying patterns.

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

Ans: The app type which is free does not add any value to the price of the  app.

```python
df['Price'].replace('Everyone',0,inplace=True)
df['Price']= df['Price'].str.replace('$','')
```

```python
df['Price']=df['Price'].astype(float)
```

```python
# Calculate average prices by app type (e.g., category or genre)
average_prices_by_type = df.groupby('Type')['Price'].mean().reset_index()
```

```python
plt.figure(figsize=(12, 6))
```

```python
app_types = average_prices_by_type['Type']
average_prices = average_prices_by_type['Price']
```

```python
# Create the bar chart
plt.bar(app_types, average_prices, color='skyblue')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
category=pd.DataFrame((df['Category']).value_counts())
category.rename(columns={'Category':'Count'},inplace=True)
```

```python
category
```

```python
# top 10
import seaborn as sns
plt.figure(figsize=(15,6))
sns.barplot(x=category.index[:10],y='Count',data=category[:10],palette='hls')
plt.title('Top 10 app categories')
plt.xticks(rotation=90)
plt.show()
```

```python
char_to_remove=['+',',']
for item in char_to_remove:
    df['Installs']=df['Installs'].str.replace(item,'')
```

```python
df['Installs']=df['Installs'].str.replace('Free','0')
```

```python
df['Installs']=df['Installs'].astype(int)
```

```python
#highest number of installs 
df_cat_installs = df.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
```

```python
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```
